```python
from statistics import correlation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns= house_price_dataset.feature_names)
```

```python
house_price_dataframe
```

```python
# Add the target column to the dataset
house_price_dataframe["price"] = house_price_dataset.target
house_price_dataframe
```

```python
# checking number of rows and column
house_price_dataframe.shape
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
correlation = house_price_dataframe.corr()
correlation
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(correlation, cbar=True, annot=True, cmap="Blues", fmt=".1f", annot_kws={"size": 8})
```

```python
X = house_price_dataframe.drop(["price"], axis = 1)
Y = house_price_dataframe["price"]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
score_1 = metrics.r2_score(Y_train, training_data_prediction)
score_2 = metrics.mean_squared_error(Y_train, training_data_prediction)
```

```python
print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
